```python
#1. What are escape characters, and how do you use them?
   
"""
Ans1:-

Escape characters are used to insert special characters or we can say which are illegal characters in a string. 
An escape character is a backslash \ followed by the character you want to insert.

Example:-

s = 'Hey, What's up?' #It will cause an SyntaxError.

To fix this error:-

s = 'Hey, What\'s up?'



There are many escape characters, some of them are given below:-


\'    Single Quote
\\    Backslash
\n    New Line
\r    Carriage Return
\t    Tab
\b    Backspace
\f    Form Feed
\ooo  Octal value
\xhh  Hex value

"""
```
